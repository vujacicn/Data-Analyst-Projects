```python
def male_female_child(passenger):
    age,sex = passenger
    
    if age < 16 :
        return 'child'
    else:
        return sex
    
```

```python
titanic_df['person'] = titanic_df[['Age','Sex']].apply(male_female_child,axis=1)
```

```python
titanic_df = titanic_df[titanic_df['Deck'] != 'T']

cat_features = ['Sex','Deck','Alone']

fig, axs = plt.subplots(ncols=3, figsize=(20, 20))
plt.subplots_adjust(wspace=0.4)

for i, feature in enumerate(cat_features):
    ax = sns.countplot(x=feature, hue='Survived', data=titanic_df, palette=["#0084ff", "#88d8b0"], ax=axs[i])
    ax.set_xlabel('{}'.format(feature), size=40, labelpad=15)
    ax.set_ylabel('Passenger Count', size=40, labelpad=15)
    ax.tick_params(axis='x', labelsize=25)
    ax.tick_params(axis='y', labelsize=25)
    ax.legend(['Not Survived', 'Survived'], loc='upper center', prop={'size': 12})
    ax.set_title('Count of Survival in {} Feature'.format(feature), size=20, y=1.05)

plt.show()
```
